```python
import requests
from bs4 import BeautifulSoup
from itertools import product

def convert_to_numeric(value):
    if '-' in value or '—' in value:
        return 0
    elif 'K' in value:
        return float(value.replace('K', '').replace(',', '')) * 1000
    elif 'M' in value:
        return float(value.replace('M', '').replace(',', '')) * 1000000
    elif 'B' in value:
        return float(value.replace('B', '').replace(',', '')) * 1000000000
    elif 'T' in value:
        return float(value.replace('T', '').replace(',', '')) * 1000000000000
    elif '%' in value:
        return float(value.replace('%', ''))
    else:
        return float(value.replace(',', ''))

def extract_data(indicator):
    try:
        # Find the indicator string in the HTML content
        indicator_string = soup.find(string=indicator)
        # Navigate to the parent tag of the indicator string
        parent_tag = indicator_string.find_parent("td")
        # Extract the value from the next sibling td tag
        value = parent_tag.find_next("td").text.strip()
        return {indicator: value}
    except AttributeError:
        print()
# Define a dictionary to store the data for each stock
stock_data = {}

for ticker, market in product(tickers, markets):
    url = f"https://www.google.com/finance/quote/{ticker}:{market}"
    response = requests.get(url)
    html_content = response.content
    # Extracted data for the current stock
    current_stock_data = {}

    # Parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')
    search_text = "We couldn't find any match for your search."
    if search_text in soup.get_text():
      continue

    # List of indicators to extract
    indicators = [
        'Cash from operations',
        'Cash from investing',
        'Cash from financing',
        'Free cash flow',
        'Cash and short-term investments',
        'Total equity',
        'Shares outstanding',
        'Price to book',
        'Return on assets',
        'Return on capital',
        'Operating expense',
        'Net profit margin',
        'Earnings per share',
        'EBITDA',
        'Effective tax rate',
    ]

    # Extract and print data for each indicator
    for indicator in indicators:
        result = extract_data(indicator)
        if result:
            key, value = result.popitem()
            convert_to_numeric(value)

            current_stock_data[key] = value

    #---------------------------------------------------------------------------------------------

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        revenue_element = soup.find('div', {'class': 'rsPbEe', 'jscontroller': 'e2jnoe', 'jsaction': 'mouseenter:tfO1Yc; focus:AHmuwe; blur:O22p3e; mouseleave:JywGue; touchstart:p6p2H; touchend:yfqBxc;mlnRJb:fLiPzd;'})

        if revenue_element:
            revenue_value = revenue_element.find_next('td', {'class': 'QXDnM'}).text

            current_stock_data['Revenue'] = revenue_value
        else:
            continue
    else:
        print(f"Failed to fetch the page. Status code: {response.status_code}")

    #---------------------------------------------------------------------------------------------

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        assets_element = soup.find('div', {'class': 'rsPbEe', 'jscontroller': 'e2jnoe', 'jsaction': 'mouseenter:tfO1Yc; focus:AHmuwe; blur:O22p3e; mouseleave:JywGue; touchstart:p6p2H; touchend:yfqBxc;mlnRJb:fLiPzd;', 'aria-describedby': 'balanceSheetTT1'})

        if assets_element:
            assets_value = assets_element.find_parent('td').find_next_sibling('td', {'class': 'QXDnM'}).text

            current_stock_data['Total Assets'] = assets_value
        else:
            continue
    else:
        print(f"Failed to fetch the page. Status code: {response.status_code}")
    #---------------------------------------------------------------------------------------------

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        liabilities_element = soup.find('div', {'class': 'rsPbEe', 'jscontroller': 'e2jnoe', 'jsaction': 'mouseenter:tfO1Yc; focus:AHmuwe; blur:O22p3e; mouseleave:JywGue; touchstart:p6p2H; touchend:yfqBxc;mlnRJb:fLiPzd;', 'aria-describedby': 'balanceSheetTT2'})

        if liabilities_element:
            liabilities_value = liabilities_element.find_next('td', {'class': 'QXDnM'}).text

            current_stock_data['Total Liabilities'] = liabilities_value
        else:
            continue
    else:
        print(f"Failed to fetch the page. Status code: {response.status_code}")
    #---------------------------------------------------------------------------------------------
    # Define the indicators
    indicators = [
        'Previous close',
        'Day range',
        'Year range',
        'Market cap',
        'P/E ratio',
        'Dividend yield'
    ]
    # Extract and print data for each indicator
    for indicator in indicators:
        section = soup.find('div', {'class': 'mfs7Fc'}, string=indicator)
        value = section.find_next('div', {'class': 'P6K39c'}).text.strip() if section else ""
        current_stock_data[indicator] = value
        if value == '':
          continue
        else:
          continue
    #---------------------------------------------------------------------------------------------

    # Add this function to calculate Bargain Score
    def calculate_bargain_score(previous_close, fifty_two_week_low, total_assets, outstanding_shares):
          score = total_assets - (previous_close * outstanding_shares)

          suffixes = ['', 'K', 'M', 'B', 'T']

          # Choose the appropriate suffix based on the magnitude of the score
          magnitude = 0
          while abs(score) >= 1000:
              magnitude += 1
              score /= 1000.0

          # Format the result with the chosen suffix
          formatted_score = f'{score:.2f}{suffixes[magnitude]}'

          return formatted_score
    def calculate_total_value_difference(previous_close, fifty_two_week_low ):
          score = previous_close - fifty_two_week_low
          suffixes = ['', 'K', 'M', 'B', 'T']

          # Choose the appropriate suffix based on the magnitude of the score
          magnitude = 0
          while abs(score) >= 1000:
              magnitude += 1
              score /= 1000.0

          # Format the result with the chosen suffix
          formatted_score = f'{score:.2f}{suffixes[magnitude]}'

          return formatted_score

    # Inside your loop for each stock, after extracting necessary data, add the following lines
    previous_close = convert_to_numeric(current_stock_data['Previous close'].replace('$', ''))
    fifty_two_week_low = convert_to_numeric(current_stock_data['Year range'].split(' - ')[0].replace('$', ''))
    outstanding_shares = convert_to_numeric(current_stock_data['Shares outstanding'])
    total_assets = convert_to_numeric(current_stock_data['Total Assets'])

    bargain_score = calculate_bargain_score(previous_close, fifty_two_week_low, total_assets, outstanding_shares)
    current_stock_data['Asset - Cap Total'] = bargain_score
    total_value_difference = calculate_total_value_difference(previous_close, fifty_two_week_low)
    current_stock_data['Price - 52wk low'] = total_value_difference
    #---------------------------------------------------------------------------------------------

    # Get Date Stamp
    from datetime import datetime

    # Get the current date and time
    current_datetime = datetime.now()

    # Format the output to include the month, day, and year
    formatted_date = current_datetime.strftime("%m/%d/%Y")
    current_stock_data['Date'] = formatted_date
    # Store the data for the current stock in the dictionary
    stock_data[f"{ticker}:{market}"] = current_stock_data
    #---------------------------------------------------------------------------------------------

    if ticker == 'ZTS':
      print(stock_data)
      # Specify the file path where you want to save the data
      file_path = f'/content/stockdata.json'

      # Write the dictionary to a JSON file
      with open(file_path, 'w') as json_file:
          json.dump(stock_data, json_file)

      print(f'Stock data has been saved to {file_path}')
    else:
      print(ticker)

```
